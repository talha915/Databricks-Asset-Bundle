```python
%sql
CREATE TABLE IF NOT EXISTS ai_lab_demo.system.execution_log
(
    run_id BIGINT,
    layer STRING,
    job_name STRING,
    status STRING,
    start_time TIMESTAMP,
    end_time TIMESTAMP,
    records BIGINT,
    error STRING
);
```

```python
%sql
CREATE TABLE IF NOT EXISTS ai_lab_demo.system.audit_log
(
    audit_id STRING,
    run_id BIGINT,
    event STRING,
    message STRING,
    timestamp TIMESTAMP
);
```

```python
%sql
CREATE TABLE IF NOT EXISTS ai_lab_demo.system.watermark
(
    pipeline_id BIGINT GENERATED ALWAYS AS IDENTITY,
    pipeline_name STRING,
    pipeline_src STRING,
    pipeline_destination STRING,
    ingestion_time TIMESTAMP,
    updated_at TIMESTAMP
);
```
